```python
import requests
import time
def scraper(id=31966071):
    item=[]
    price=[]
    trial=0
    steps=30
    num=steps-1
    start=0
    user_agent = "Mozilla/5.0 (Windows NT 10.0; WOW64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/79.0.3945.88 Safari/537.37"
    end=0
    print('{:^8}{:^100}{:^9}{:^10}'.format('No','Nama barang','stok','harga'))
    while num==steps-1 and trial<10:
        # r=requests.get("https://shopee.co.id/api/v4/search/search_items?by=pop&entry_point=ShopByPDP&limit=30&match_id="+str(id)+"&newest="+str(start)+"&order=desc&page_type=shop&scenario=PAGE_OTHERS&version=2",headers={'User-Agent': user_agent})
        r=requests.get("https://shopee.co.id/api/v4/shop/search_items?filter_sold_out=1&limit=30&offset=0&order=desc&shopid="+str(id)+"&sort_by=pop&use_case=4",headers={'User-Agent': user_agent})
        try:
            for num,x in enumerate(r.json()['items']):
                item.append([x['item_basic']['name']])
                price.append(int(x['item_basic']['price_max'])/100000)
                print('{:^8}{:^100}{:^9}{:^10}'.format(num+end+1,x['item_basic']['name'],x['item_basic']['stock'],str(int(x['item_basic']['price_max'])/100000)))
            start+=steps
            end=num+end+1
            time.sleep(5)
            trial=0
        except:
            trial+=1
            print(r)
            time.sleep(6)
    return (item,price)
```
