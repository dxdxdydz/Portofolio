```python
import nltk
import pandas as pd
import numpy as np
from nltk.corpus import stopwords
import string
import re
from collections import Counter, OrderedDict
from matplotlib import pyplot as plt
```

```python
def unique(sequence,):
    seen = set()
    return [x for x in sequence if not (x in seen or seen.add(x))]  

def tokenize(text,frequent_words=[]):
    text=text.lower()
    text = "".join([char for char in text if char not in string.punctuation])
    words=[word for word in nltk.word_tokenize(text) if word not in stopwords.words('indonesian')]
    words=[word for word in words if word not in stopwords.words('english')]
    if frequent_words==[]:
        pass
    else:
        words=[word for word in nltk.word_tokenize(text) if word not in frequent_words]
    words=[u_split(word) for word in words]
    tokens=[token for lists in words for token in lists]    
    return unique(tokens)
def frequent_words(a_item,b_item):
    dictionary1=[]
    for a in a_item:
        dictionary1+=tokenize(*a)
    dictionary2=[]
    for a in b_item:
        dictionary2+=tokenize(*a)
    return set(list(OrderedDict(Counter(dictionary1).most_common()))[:40]).intersection(set(list(OrderedDict(Counter(dictionary1).most_common()))[:40]))

def ppcs(text):
    return tokenize(unique(u_split(text)))

def u_split(text):
    k=re.match(r'(\d*)(\w*)',text)
    if k[1].isdigit() and k[2].isalpha():
        return [k[1],k[2]]
    else :
        return [text]
```
